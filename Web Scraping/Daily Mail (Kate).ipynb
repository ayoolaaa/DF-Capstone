# WEBSCRAPING

```python
kate_dailymail = pd.DataFrame(columns=['news_source','date','headline', 'url'])
```

```python
def scrap_article_urls(offset):
    
    url = f'https://www.dailymail.co.uk/home/search.html?offset={offset}&size=50&sel=site&searchPhrase=kate+middleton&sort=recent&type=article&days=all'
    
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    
    article_urls = [x.a.get('href') for x in soup.findAll('h3', class_='sch-res-title')]
    
    headlines = [x.a.text for x in soup.findAll('h3', class_='sch-res-title')]
    
    df = pd.DataFrame(article_urls, columns=['url'])
    df['headline'] = headlines
    
    return df
```

```python
offset=0

for offset in range(0,15100,50):
    df = scrap_article_urls(offset)
    kate_dailymail = kate_dailymail.append(df)
    
kate_dailymail.reset_index()
```

# WEBSCRAPING FOR PUBLICATION DATE

```python
def get_article_details(url):
    # URL from internet
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    
    # Date Published
    date = soup.find('span', class_='article-timestamp article-timestamp-published').time.text
    
    return date
```

```python
for i in range(0, len(kate_dailymail)):
    
    url = kate_dailymail['url'].iloc[i]
    news_source = kate_dailymail['news_source'].iloc[i]
    headline = kate_dailymail['headline'].iloc[i]
    date = get_article_details(url)
    kate_dailymail.iloc[i] = news_source, date, headline, url
```
