```python
def scrap_article_urls(offset):
    
    url = f'https://www.dailymail.co.uk/home/search.html?offset={offset}&size=50&sel=site&searchPhrase=meghan+markle&sort=recent&type=article&type=video&type=permabox&days=all'

    
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    
    article_urls = [x.a.get('href') for x in soup.findAll('h3', class_='sch-res-title')]
    
    headlines = [x.a.text for x in soup.findAll('h3', class_='sch-res-title')]
    
    df = pd.DataFrame(article_urls, columns=['url'])
    df['headline'] = headlines
    
    return df
```

```python
offset=0

for offset in range(0,12700,50):
    df = scrap_article_urls(offset)
    meg_articles = meg_articles.append(df)
    
meg_articles.reset_index()
```

# WEBSCRAPING FOR PUBLICATION DATE

```python
site = 'https://www.dailymail.co.uk/news/article-9711525/Meghan-Markle-ready-brutal-showdown-Buckingham-Palace.html'
result = requests.get(site)
soup = BeautifulSoup(result.text, 'html.parser')
```

```python
soup.find('span', class_='article-timestamp article-timestamp-published').time.text
```

```python
def get_article_details(url):
    # URL from internet
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    
    # Date Published
    date = soup.find('span', class_='article-timestamp article-timestamp-published').time.text
    
    return date
```

```python
for i in range(0, len(meg_dailymail)):
    
    url = meg_dailymail['url'].iloc[i]
    news_source = meg_dailymail['news_source'].iloc[i]
    headline = meg_dailymail['headline'].iloc[i]
    date = get_article_details(url)
    meg_dailymail.iloc[i] = news_source, date, headline, url
```
